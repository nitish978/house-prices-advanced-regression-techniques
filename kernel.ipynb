```python
df=pd.read_csv("train.csv")
df=df.fillna(method='ffill')
df1=pd.read_csv("test.csv")
df1=df1.fillna(method='ffill')
```

```python
new_df=df.select_dtypes(include=['object'])
col=new_df.columns
```

```python
def convert_to_num(df,df1,X):
  for x in X :
     x1=set(df[x])
     x1=list(x1)
     j=1
     for i in x1:
        df=df.replace(i,j)
        df1=df1.replace(i,j)
        j=j+1
  return df,df1      
```

```python
df,df1=convert_to_num(df,df1,col)
```

```python
df=df.sample(frac=1)
col=df.columns
train=df[col[1:len(col)-1]]
```

```python
labels=df[col[-1]]
```

```python
col=df1.columns
test=df1[col[1:]]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(train)
train=transformer.transform(train)
transformer1 = Normalizer().fit(test)
test=transformer1.transform(test)
```

```python
reg = LinearRegression().fit(train,labels)
```

```python
pre=reg.predict(test)
pre
```
